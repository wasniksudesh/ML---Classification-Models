```
# Linear Regression Closed Form on Adult All feature dataset
import numpy as np
import pandas as pd
f=open("/content/all-features-encoded-adult.csv")
df=pd.read_csv(f);
ran=[];
minn=[];
for i in df:
    l=df[i].tolist()
    a,b=min(l),max(l)
    minn.append(a)
    ran.append(b-a)

f=open("/content/all-features-encoded-adult.csv")

data=[]
label=[]
temp=0;
for i in f:
    if(temp==0):
        temp=1
        continue
    l=i.split(",")
    l[-1]=float(l[-1][0])

    l=[round((float(l[i])-minn[i])/ran[i],2) for i in range(len(l))]
    label.append(((l[2])))
    l=l[:2]+l[3:]
    l=[float(j) for j in l];

    l=[1]+l[1:]
    data.append(l)
n=len(data)
m=len(data[0])
print(n)
base=data;

test=data[25000:]
train=data[:25000];
test_label=label[25000:]
train_label=label[:25000]
n=len(train)
print(n)
print(label)
```

```
## Getting Data and Initializing --  Medical Classification
import numpy as np
f=open("/content/Medical_data.csv")
data=[]
label=[]
temp=0;
for i in f:
    if(temp==0):
        temp=1
        continue
    l=i.split(",")
    label.append(l[0])
    l=l[1:]
    l=[float(j) for j in l];
    data.append(l)
n=len(data);
for i in range(n):
    if(label[i]=='HEALTHY'):
        label[i]=0
    else:
        label[i]=1
train=data;
train_label=label
print(n)
print(train)
# test=data
# train=data[:25000];
# test_label=label[25000:]
# train_label=label[:25000]
# n=len(train)
# print(n)
# print(train)
```

```
##Using the formula to calulate 
nptrain=np.array(train)
d=len(train[0])
xtx=np.dot(nptrain.T,nptrain)
xtx=xtx+ 1*np.identity(d)
inv=np.linalg.inv(xtx);
prod=np.dot(inv,nptrain.T);
weights=np.dot(prod,train_label);
print(weights)
```

```
##Using the formula to calulate 
nptrain=np.array(train)
exc=[]
for lam in range(1,1000000,10000):
    d=len(train[0])
    xtx=np.dot(nptrain.T,nptrain)
    xtx=xtx+ lam*np.identity(d)
    inv=np.linalg.inv(xtx);
    prod=np.dot(inv,nptrain.T);
    weights=np.dot(prod,train_label);
    #     print(weights)
    #Testing -  Binary Classification on Adult
    ans=[]
    peak=[]
    for i in range(len(test)):
        q=np.dot(np.array(test[i]),weights);
        ans.append(round(q,2))
        if(abs(1-ans[-1])<abs(0-ans[-1])):
            peak.append(1)
        else:
            peak.append(0)
        ans.append(round(abs(q-test_label[i]),2));
    #     print(ans)

#     print(test_label)
#     print(peak)
#     count=0;
#     for i in range(len(test)):
#         if(peak[i]==test_label[i]):
#             count+=1;
#     print(count/len(test))
    # from sklearn import metrics
    exc.append((np.linalg.norm(ans)/(len(test)**(1/2))))
    print(exc[-1])
```
