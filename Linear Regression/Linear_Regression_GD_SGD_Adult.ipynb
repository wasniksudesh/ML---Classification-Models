```
# Linear Regression Gradient Descent on Adult All feature dataset
import numpy as np

import pandas as pd

f=open("/content/all-features-encoded-adult.csv")
df=pd.read_csv(f);
ran=[];
minn=[];
for i in df:
    l=df[i].tolist()
    a,b=min(l),max(l)
    minn.append(a)
    ran.append(b-a)

f=open("/content/all-features-encoded-adult.csv")
data=[]
label=[]




temp=0;
for i in f:
    if(temp==0):
        temp=1
        continue
    l=i.split(",")
    l=[round((float(l[i])-minn[i])/ran[i],2) for i in range(len(l))]
    label.append((float(l[2])))
    l=l[:2]+l[3:]
#     l[-1]=float(l[-1][0])
    l=[1]+l[1:]
    l=[float(j) for j in l];
    data.append(l)
n=len(data)
m=len(data[0])
print(n)
base=data;
test=data[25000:]
train=data[:25000];
test_label=label[25000:]
train_label=label[:25000]
n=len(train)
print(n)
d=len(train[0])
print(train)


# Predictor fucntion
def h(theta,x):
    theta=np.array(theta)
    x=np.array(x).T
    l=np.dot(theta,x);
    return l;
```

```
# Applying theta vector -- Batch GD
theta=[0.01]*d

epochs=100
alpha=0.0001
mm=[0]*d;
print(mm)

for i in range(epochs):
    mm=[0]*d;
    for j in range(n):
        mm+= (h(theta,train[j])-train_label[j])*np.array(train[j]);
    theta=theta- alpha*(mm + -1*np.array(theta))*(1/n);
    
print(theta)

#testing
peak=[]
ans=[]
for i in range(len(test)):
    ans.append(int(h(theta,test[i])))
    peak.append(float(h(theta,test[i])-test_label[i]))
print(ans)
print(test_label)
print(np.linalg.norm(np.array(peak))/(len(test)**(1/2)))

    
```
